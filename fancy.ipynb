```python
import pandas as pd
```

```python
fancy = pd.read_csv("data/Fancy - approach group.csv")
```

```python
dives = []

for _, row in fancy.iterrows():
    new_rows = pd.DataFrame.from_dict(
        {
            "label": row["label"],
            "som": [row["som_1"], row["som_2"], row["som_3"], row["som_4"], row["som_5"]],
            "twist": [row["twist_1"], row["twist_2"], row["twist_3"], row["twist_4"], row["twist_6"]],
            "str": [row["str_1"], row["str_2"], row["str_3"], row["str_4"], row["str_5"]],
            "pike": [row["pike_1"], row["pike_2"], row["pike_3"], row["pike_4"], row["pike_5"]],
            "tuck": [row["tuck_1"], row["tuck_2"], row["tuck_3"], row["tuck_4"], row["tuck_5"]],
            "free": [row["free_1"], row["free_2"], row["free_3"], row["free_4"], row["free_5"]],
        }
    )
    dives.append(new_rows)

all_dives = pd.concat(dives)

all_dives = all_dives.loc[~(all_dives["som"].isna() & all_dives["twist"].isna() & all_dives["str"].isna() & all_dives["pike"].isna() & all_dives["tuck"].isna() & all_dives["tuck"].isna())]
all_dives = all_dives.reset_index(drop=True)
all_dives["image"] = all_dives.index.map(lambda i: f"https://jsofra.github.io/fancy/images/dive_cards/dive_card_{i}.jpg")

split_label = all_dives["label"].str.split(" ")

all_dives["group"] = split_label.map(lambda x: x[0])
all_dives["sub_group"] = split_label.map(lambda x: x[1] if len(x) > 1 else "").str.replace("(", "").str.replace(")", "")

all_dives = all_dives[["group", "sub_group", "som", "twist", "str", "pike", "tuck", "free",	"image"]]
```

```python
all_dives.dropna(subset=["free"], axis=0)
```
